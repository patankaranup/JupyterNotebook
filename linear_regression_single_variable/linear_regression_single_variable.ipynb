```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
df.head(3)
```

```python
%matplotlib inline
```

```python
# Scatter plot requires x and y as arguments
plt.xlabel("Area of House")
plt.ylabel("Price of House")
plt.scatter(df.area,df.price)
```

```python
linear_regression = linear_model.LinearRegression()
```

```python
# this expects 2d array of independent varaible 
# convert the dataframe into 2d array
X = df['area'].values.reshape(-1, 1) #reshape because it has only single feature
y = df['price'].values
# print(type(X))
linear_regression.fit(X,y)
```

```python
# Best fit line graph
%matplotlib inline
plt.xlabel('area',fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,linear_regression.predict(X))
```

```python
linear_regression.predict([[3500]])
```

```python
# slope or gradient of the line
linear_regression.coef_
```

```python
# y_intercept of the line
linear_regression.intercept_
```

```python
# Slope of a line is y = mx+c
# y = mx+c

pred_val = 135.78767123*3500+180616.43835616432
print(pred_val)
```

```python
areas = pd.read_csv('areas.csv')
```

```python
areas.head(3)
```

```python
# predict the prices expects an array
areas_array = areas.values
predicted_prices = linear_regression.predict(areas_array)
```

```python
areas['Predicted_Prices'] = predicted_prices
```
