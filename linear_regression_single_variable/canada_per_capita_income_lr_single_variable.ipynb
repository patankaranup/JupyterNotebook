```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('./canada_per_capita_income.csv')
```

```python
df.head(3)
```

```python
%matplotlib inline
```

```python
# Scatter plot requires x and y as arguments
plt.xlabel("Year")
plt.ylabel("Income in Canada")
plt.scatter(df.year,df['per capita income (US$)'])
```

```python
linear_regression = linear_model.LinearRegression()
```

```python
# this expects 2d array of independent varaible 
# convert the dataframe into 2d array
X = df['year'].values.reshape(-1, 1) #reshape because it has only single feature
y = df['per capita income (US$)'].values
# print(type(X))
linear_regression.fit(X,y)
```

```python
# Best fit line graph
%matplotlib inline
plt.xlabel('Year',fontsize=20)
plt.ylabel('Income', fontsize=20)
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='+')
plt.plot(df.year,linear_regression.predict(X))
```
