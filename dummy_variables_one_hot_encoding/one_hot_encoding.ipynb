```python
import pandas as pd
df = pd.read_csv("homeprices.csv")
df
```

```python
# get dummy variables for town column
pd.get_dummies(df[['town']])
```

```python
dummies = pd.get_dummies(df[['town']])
dummies
```

```python
# merger dataframe on col
merged_df = pd.concat([df,dummies],axis='columns')
merged_df
```

```python
# remove town column
final_df = merged_df.drop(['town','town_west windsor'],axis='columns')
```

```python
final_df
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final_df.drop(['price'],axis='columns').values
y = final_df['price'].values
```

```python
X
```

```python
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.predict([[2800,1,0]])
```

```python
# calculate the accuracy of the model
model.score(X,y)
```
